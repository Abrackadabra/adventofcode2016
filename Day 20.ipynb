```python
ranges = []
for line in lines:
    a, b = list(map(int, line.split('-')))
    ranges.append((a, b))
ranges = sorted(ranges)
```

```python
i = 0
t = 0
c = 0

while i < len(ranges) and t <= 4294967295:
    r = ranges[i]
    if r[1] < t:
        i += 1
        continue
    if r[0] <= t <= r[1]:
        t = r[1] + 1
        continue
    if t < r[0]:
        c += r[0] - t
        t = r[0]
        continue
    
t, c
```
