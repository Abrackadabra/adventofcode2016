```python
M = array([[-1 if j == '#' else (0 if j == '.' else int(j) + 1) for j in line] for line in lines])
```

```python
points = {}
coords = list(itertools.product(range(M.shape[0]), range(M.shape[1])))
```

```python
for i, j in coords:
    if M[i, j] > 0:
        points[M[i, j]] = (i, j)
```

```python
dirs = [
    [0, 1], [1, 0], [0, -1], [-1, 0]
]
```

```python
def neighbours(point):
    for d in dirs:
        x = point[0] + d[0]
        y = point[1] + d[1]
        
        if x < 0 or x >= M.shape[0] or y < 0 or y >= M.shape[1]:
            continue
        
        if M[x, y] < 0:
            continue
        
        yield (x, y)
```

```python
dist = {}
for pointi, point in points.items():
    q = collections.deque()
    visited = {}
    q.append(point)
    visited[point] = 0
    
    while q:
        p = q.popleft()
        
        for friend in neighbours(p):
            if friend in visited:
                continue
            visited[friend] = visited[p] + 1
            q.append(friend)
            
    d = {i: visited[point] for i, point in points.items()}
    dist[pointi] = d
```

```python
minL = 10 ** 9
minP = None
for path in itertools.permutations(list(range(1, 9))):
    if path[0] != 1:
        continue
    t = 0
    for i in range(len(path) - 1):
        t += dist[path[i]][path[i + 1]]
    if t < minL:
        minL = t
        minP = path
```
