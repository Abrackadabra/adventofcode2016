```python
s = 'abcdefgh'
ms = lambda s: ''.join(s)
for line in lines:
    print(s, line)
    for a, b in re.findall(r'swap position (\d+) with position (\d+)', line):
        a = int(a)
        b = int(b)
        s = list(s)
        t = s[a]
        s[a] = s[b]
        s[b] = t
        s = ''.join(s)
    for a, b in re.findall(r'swap letter (.) with letter (.)', line):
        for i in range(len(s)):
            if s[i] == a:
                s = s[:i] + b + s[i+1:]
            elif s[i] == b:
                s = s[:i] + a + s[i+1:]
    for a, b in re.findall(r'rotate (\w+) (\d+) steps', line):
        b = int(b)
        for i in range(b):
            if a == 'left':
                s = s[1:] + s[0]
            else:
                s = s[-1] + s[:-1]
    for a in re.findall(r'rotate based on position of letter (.)', line):
        i = line.find(a)
        c = i + 1
        if i >= 4:
            c += 1
        for i in range(c):
            s = s[-1] + s[:-1]
    for a, b in re.findall(r'reverse positions (\d+) through (\d+)', line):
        a, b = int(a), int(b)
        s = s[:a] + ms(s[a:b+1][::-1]) + s[b+1:]
    for a, b in re.findall(r'move position (\d+) to position (\d+)', line):
        a, b = int(a), int(b)
        c = s[a]
        s = s[:a] + s[a+1:]
        s = s[:b] + c + s[b:]
```
